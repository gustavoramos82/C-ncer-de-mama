# Verificando a perfomace dos modelo sem otimização

```python
# Importando as blibiotecas

import numpy as np 
import pandas as pd 

from sklearn.svm import SVC 
from sklearn.naive_bayes import GaussianNB 
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import cross_validate, KFold
from imblearn.under_sampling import RandomUnderSampler
from sklearn.linear_model import LogisticRegression
```

```python
breast = pd.read_csv('data.csv')
```

```python
breast.head()
```

```python
breast.drop(['id','Unnamed: 32'],axis=1,inplace=True)
```

```python
breast.head()
```

```python
breast['diagnosis'] = breast['diagnosis'].map({'M':1,'B':0})
```

```python
breast.head()
```

```python
# Separando os dados e atributos e target
x = breast.drop(['diagnosis'],axis=1).values
y = breast.diagnosis.values

# Normalizando os dados
x_std = StandardScaler().fit_transform(x)
```

```python
# usando o metos undersampling para tornar os dados menos enviesados e comparar as metricas
nm = RandomUnderSampler(random_state = 10)

x_nm, y_nm = nm.fit_resample(x_std, y)
```

```python
svm = SVC(probability=True)
knn = KNeighborsClassifier()
tree = DecisionTreeClassifier()
na_ba = GaussianNB()
lg = LogisticRegression()
```

```python
# colocando o kfold
kfold = KFold(n_splits=7)
```

#### Aplicando o svm

```python
nome_metricas = ['accuracy', 'precision', 'recall','f1']

metricas_ran = cross_validate(svm,x_std, y, cv=kfold, scoring=nome_metricas)

for met in metricas_ran:
  print(f'-{met}')
  print(f"-- {metricas_ran[met]}")
  media = np.mean(metricas_ran[met])
  desvio = np.std(metricas_ran[met])
  print(f'Média do {met}: {media}')
  print(f'Desvio {desvio}')
  print(f'Intervalo [{(media-(2*desvio)):.3f},{(media+(2*desvio)):.3f}]')
  print('-*-'*20)
```

```python
metricas_ran_u = cross_validate(svm,x_nm, y_nm, cv=kfold, scoring=nome_metricas)

for met in metricas_ran_u:
  print(f'-{met}')
  print(f"-- {metricas_ran_u[met]}")
  media = np.mean(metricas_ran_u[met])
  desvio = np.std(metricas_ran_u[met])
  print(f'Média do {met}: {media}')
  print(f'Desvio {desvio}')
  print(f'Intervalo [{(media-(2*desvio)):.3f},{(media+(2*desvio)):.3f}]')
  print('-*-'*20)
```

Tivemos os seguintes resultados com svm:

- sem *undersampling*:
  - acuracia: 0,973
  - precisão: 0,966
  - recall: 0,958
  - f1: 0,962

- com *undersampling*:
  - acuracia: 0,964
  - precisão: 0,563
  - recall: 0.549
  - f1: 556
